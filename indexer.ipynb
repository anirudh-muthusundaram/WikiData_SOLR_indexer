## Apache SOLR Indexing

### Import Libraries

```python
import os
import pysolr
import csv
import requests
import pandas as pd
import pickle
```

## Assign Core Name and IP

```python
# Assign the Core Name 
# Assign the IP address
CORE_NAME = "IRF23P1"
VM_IP = "35.245.110.213:8983"
```

```python
#Load the preprocessed data
data = 'preprocessed_text.csv'
link_SOLR = f'http://{VM_IP}/solr/{CORE_NAME}'
```

```python
#initialize the solr 
solr = pysolr.Solr(link_SOLR)
```

## Indexing the CSV

```python
#upload all of the documents in iteration.
with open(data, mode='r', encoding='utf-8') as file:
    docs_input = csv.DictReader(file)
    
    # Map all the DataFrames based on the Schema.
    for docs in docs_input:
        document = {
            'revision_id': docs['Revision ID'],
            'Title': docs['Title'],
            'summary': docs['Summary'],
            'url': docs['URL'],
            'topic': docs['Topic'],
        }
        # upload the documents
        output = solr.add([document])
        # print log
        print(output)
```
